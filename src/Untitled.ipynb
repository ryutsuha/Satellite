```python
import time
import numpy as np
import os
import csv
import datetime


runtime   = time.time()

TOTAL_POWER = 5
num_of_beam = 8
BEAM_STEP   = 5
power_list  = []

POWER_GRADATION = num_of_beam * BEAM_STEP
power_width     = TOTAL_POWER / POWER_GRADATION


print("POWER_GRADATION : ", POWER_GRADATION)
print("power_width     : ", power_width)

def printlen() :
    if len(power_list) > 1 and len(power_list) % 10000 == 0 :
        print(len(power_list), (time.time() - runtime)/60)

def decision_power() :
    def dfs(A):

        # 数列の長さが num_of_beam に達したら打ち切り
        if len(A) == num_of_beam:
            # 処理
            if sum(A) == TOTAL_POWER :
                power_list.append(A.copy())
                printlen()
            return
        
        for v in range(POWER_GRADATION+1):
            A.append(power_width * v if v != 0 else 0)
            dfs(A)
            A.pop()

    dfs([])

    print(len(power_list), time.time() - runtime)
    # for _ in power_list :
    #     print(_)

    return power_list


def result_output():

    filename = "database\\decision_power_" + str(POWER_GRADATION) + "_" + datetime.datetime.now().strftime("%m%d-%H%M%S") + ".csv"

    with open(filename, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(range(num_of_beam))
        for _ in power_list :
            # print(_)
            writer.writerow(_)

power_list = decision_power()
result_output()
```
